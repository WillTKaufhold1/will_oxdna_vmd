```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# we want to someone output the energy per nucleotide in lots of senses
# we therefore need to be able to read in a file, interpret each of the energies
# and then appropriately pool energies.

with open("../10-30-ox-test/test-energies.dat") as f : data = f.readlines()

all_data = []

local = []

for i in data:
    if i[:2] == "#T":
        if local != []: 
            all_data.append(np.array(local))
            local = []
    elif i[:2] == "#i":
        pass
    else:
        local.append(list(map(float,i.replace('\n','').split())))

all_data = np.array(all_data)

Max = max(all_data[0][:,1])

energy_dict = {}

for frame in all_data:
    for line in frame:
        if line[0]+1 not in energy_dict:
            energy_dict[line[0]+1] = np.zeros(len(line)-2)
        if line[1]+1 not in energy_dict:
            energy_dict[line[1]+1] = np.zeros(len(line)-2)
        energy_dict[line[0]+1] += np.array(line[2:]) / len(all_data)
        energy_dict[line[1]+1] += np.array(line[2:]) / len(all_data)
```

```python
#id1 id2 FENE BEXC STCK NEXC HB CRSTCK CXSTCK DH total, t = 10000
```

```python
energy = []
```

```python
for key in np.arange(1,len(energy_dict)+1):
    energy.append(energy_dict[key])
```
